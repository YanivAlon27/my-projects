```python
class Chips():
    
    def __init__(self,chips):
        self.chips = chips
    
    def wins(self,win):
        self.chips = self.chips + win
        print(f"Added {win} to the chips")
        print("Player current chips are: " + str(self.chips))
    def lost(self,lose):
        
        if self.chips >= lose:
            self.chips = self.chips - lose
            print("Your chips now is: " + str(self.chips))
        else:
            print("Game over")
    
    def __str__(self):
        return "Player current chips are: " + str(self.chips)
    
```

```python
def ace(lst,sum_of_cards):
    for word in lst:
        if sum_of_cards > 21 and word[0] == "a":
            sum_of_cards = sum_of_cards - 10
    return(sum_of_cards)

```
